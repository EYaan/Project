```python
y15_16["count_score"] = 2
y15_16["count_gdp"] = 2
y15_16["count_support"] = 2
y15_16["count_life"] = 2
y15_16["count_freedom"] = 2
y15_16["count_generosity"] = 2
y15_16["count_curruption"] = 2

# 평균값을 구하기 위한 함수에 사용 될 숫자
```

```python
y15_16.head()
```

```python
def sum_data(data):
    
    data[["count_score",'count_gdp',"count_support","count_life","count_freedom","count_generosity",
          "count_curruption"]] = data[["count_score",'count_gdp',"count_support","count_life","count_freedom",
          "count_generosity","count_curruption"]].fillna(2)
    
    # 처음엔 2로 다 넣어 줬지만 다음 함수 사용을 위해 결측값을 2로 넣는 작업
    
    
    
    for i in data.index: # 함수에 들어온 데이터프레임의 인덱스 만큼 반복
     
        if data.isnull()['Ladder score_x'][i] and not(data.isnull()['Ladder score_y'][i]):
            data['Ladder score_x'][i] = data['Ladder score_y'][i]
            # 처음 데이터 프레임에 없는 나라의 데이터가 들어있을 경우
            # nan 값에 Ladder score_y 데이터를 Ladder score_x 로 값을 옮겨오는 작업
            # 평균을 구하지않고 대입만 했기 때문에 카운트 증가 x
        elif not(data.isnull()['Ladder score_x'][i]) and not(data.isnull()['Ladder score_y'][i]):
            data['Ladder score_x'][i] = (data['Ladder score_y'][i]+data['Ladder score_x'][i]*((data['count_score'][i])-1))/data['count_score'][i]
            data['count_score'][i]+=1
            # 둘다 값이 들어 있을 경우 합하여 나누는 작업
            # 몇개의 데이터의 평균값이 Ladder score_x에 들어있을 수 있기 때문에
            # (카운트-1)로 곱하고 Ladder score_y와 합한후 다시 카운트로 나누는 작업
            # 이하 반복
            
        if data.isnull()['Logged GDP per capita_x'][i] and not(data.isnull()['Logged GDP per capita_y'][i]):
            data['Logged GDP per capita_x'][i] = data['Logged GDP per capita_y'][i]
        elif not(data.isnull()['Logged GDP per capita_x'][i]) and not(data.isnull()['Logged GDP per capita_y'][i]):
            data['Logged GDP per capita_x'][i] = (data['Logged GDP per capita_y'][i]+data['Logged GDP per capita_x'][i]*((data['count_gdp'][i])-1))/data['count_gdp'][i]
            data['count_gdp'][i]+=1

        if data.isnull()['Social support_x'][i] and not(data.isnull()['Social support_y'][i]):
            data['Social support_x'][i] = data['Social support_y'][i]
        elif not(data.isnull()['Social support_x'][i]) and not(data.isnull()['Social support_y'][i]):
            data['Social support_x'][i] = (data['Social support_y'][i]+data['Social support_x'][i]*((data['count_support'][i])-1))/data['count_support'][i]
            data['count_support'][i]+=1
        
        if data.isnull()['Healthy life expectancy_x'][i] and not(data.isnull()['Healthy life expectancy_y'][i]):
            data['Healthy life expectancy_x'][i] = data['Healthy life expectancy_y'][i]
        elif not(data.isnull()['Healthy life expectancy_x'][i]) and not(data.isnull()['Healthy life expectancy_y'][i]):
            data['Healthy life expectancy_x'][i] = (data['Healthy life expectancy_y'][i]+data['Healthy life expectancy_x'][i]*((data['count_life'][i])-1))/data['count_life'][i]
            data['count_life'][i]+=1
            
        if data.isnull()['Freedom to make life choices_x'][i] and not(data.isnull()['Freedom to make life choices_y'][i]):
            data['Freedom to make life choices_x'][i] = data['Freedom to make life choices_y'][i]
        elif not(data.isnull()['Freedom to make life choices_x'][i]) and not(data.isnull()['Freedom to make life choices_y'][i]):
            data['Freedom to make life choices_x'][i] = (data['Freedom to make life choices_y'][i]+data['Freedom to make life choices_x'][i]*((data['count_freedom'][i])-1))/data['count_freedom'][i]
            data['count_freedom'][i]+=1

        if data.isnull()['Generosity_x'][i] and not(data.isnull()['Generosity_y'][i]):
            data['Generosity_x'][i] = data['Generosity_y'][i]
        elif not(data.isnull()['Generosity_x'][i]) and not(data.isnull()['Generosity_y'][i]):
            data['Generosity_x'][i] = (data['Generosity_y'][i]+data['Generosity_x'][i]*((data['count_generosity'][i])-1))/data['count_generosity'][i]
            data['count_generosity'][i] +=1
            
        if data.isnull()['Perceptions of corruption_x'][i] and not(data.isnull()['Perceptions of corruption_y'][i]):
            data['Perceptions of corruption_x'][i] = data['Perceptions of corruption_y'][i]
        elif not(data.isnull()['Perceptions of corruption_x'][i]) and not(data.isnull()['Perceptions of corruption_y'][i]):
            data['Perceptions of corruption_x'][i] = (data['Perceptions of corruption_y'][i]+data['Perceptions of corruption_x'][i]*((data['count_curruption'][i])-1))/data['count_curruption'][i]
            data['count_curruption'][i] +=1
            
    data.drop(['Ladder score_y','Logged GDP per capita_y','Social support_y','Healthy life expectancy_y','Freedom to make life choices_y','Generosity_y','Perceptions of corruption_y'], axis=1, inplace = True)
    # 쓸모없어진 데이터 제거

    return data
```
